```python
# clean 'votings' column
df1['votings'] = df1['votings'].str.extract(r'(\d+)').astype(float).astype(int)
```

```python
df1
```

```python
# Function to convert voting_counts safely
import re
def convert_votes(value):
    if pd.isna(value) or value.strip()== "": # handle NaN or empty values
        return 0
        
    value = re.sub(r"[^\dKM.]","",value)  #remove brackets

    if 'K' in value:
        return float(value.replace('K','').strip())*1000
    elif 'M' in value:
        return float(value.replace('M','').strip())*1000000
    else:
        try:
            return float(value.strip())
        except ValueError:
            return 0 # return 0 if conversion fails
        
df1["voting_counts"]= df1["voting_counts"].astype(str).apply(convert_votes)
df1
```
